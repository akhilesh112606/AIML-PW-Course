```python
# Question 1
def even_sum(temp_list):
  sum = 0
  for num in temp_list:
    if num%2==0:
      sum+=num
  return sum

temp_list = [1,2,3,4,5]
print(even_sum(temp_list))
```

```python
# Question 3
def sq_list(ele):
  new_list = []
  for num in ele:
    new_list.append(num**2)
  return new_list

temp_list = [1,2,3,4,5]
print(sq_list(temp_list))
```

```python
# Question 5
import math
def prime_check(n):
  i = 2
  while i<= math.sqrt(n):
    if n%i==0:
      return False
    i+=1
  return True

print(prime_check(6))
```

```python
# Question 6
def power(n):
  for i in range(n+1):
    yield 2**i
iter = power(5)
for i in range(5):
  print(next(iter))
```
